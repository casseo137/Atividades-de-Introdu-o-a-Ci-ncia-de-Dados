## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

```python
import pandas as pd

filmes = pd.read_csv("movies.csv")
filmes
```

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI

genero = filmes.groupby("Major Genre").mean().sort_values(by="US DVD Sales", ascending=False)[["US DVD Sales"]]
genero.iloc[0:1]
```

### Qual o diretor dirigiu mais filmes?

```python
# ADICIONE O SEU CÓDIGO AQUI

diretor = filmes["Director"].value_counts()
diretor.iloc[0:1]
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI

valor = filmes.sort_values("Title")[["Title", "Worldwide Gross", "Production Budget"]]
valor["Lucro"] = valor["Worldwide Gross"] - valor["Production Budget"]
valor_final = valor.sort_values("Lucro", ascending=False)
valor_final.iloc[0:21]
```
